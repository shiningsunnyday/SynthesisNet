```python
import pickle
data = pickle.load(open('SynthesisNet/results/viz/skeletons.pkl', 'rb'))
sts = list(data)
```

```python
from synnet.utils.data_utils import Skeleton

indices = {}
for i in range(len(sts)):
    sk = Skeleton(sts[i], i)
    if sk.rxns.sum() <= 6:
        indices[sk.rxns.sum()] = indices.get(sk.rxns.sum(), []) + [i]
```

```python
from pathlib import Path
import shutil
import os
for max_depth in [5,4,2,1]:
    for suffix in ['train','valid','test']:
        orig_folder = f'SynthesisNet/data/gnn_featurized_rxn_target_down_interm_postorder_max_depth=6_{suffix}'
        p = Path(orig_folder)
        for depth in range(max_depth, 0, -1):
            for ind in indices[depth]:
                files = p.rglob(f'{ind}_*')                
                for file in files:
                    dest_folder = f'SynthesisNet/data/gnn_featurized_rxn_target_down_interm_postorder_max_depth={max_depth}_{suffix}'
                    os.makedirs(dest_folder, exist_ok=True)
                    dest_path = os.path.join(dest_folder, file.name)
                    shutil.copyfile(file, dest_path)
```
